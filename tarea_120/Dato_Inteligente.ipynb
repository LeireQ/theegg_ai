```python
#1. Promedio tasa de fertilidad global
df = df.dropna() #elimina las filas que contienen datos nulos
df['Tasa Fertilidad'] = df['Tasa Fertilidad'].astype(float)
promedio_fertilidad = df['Tasa Fertilidad'].mean()
promedio_fertilidad
```

```python
#2. Edad promedio más alta
df['Edad Promedio'] = df['Edad Promedio'].astype(float)
edad_promedio_max = df['Edad Promedio'].max()
edad_promedio_max
```

```python
# País con edad promedio más alta
indice_promedio_max = df['Edad Promedio'].idxmax()
pais_promedio_max = df.iloc[indice_promedio_max]['Pais']
pais_promedio_max
```

```python
#2. Edad promedio más baja
df['Edad Promedio'] = df['Edad Promedio'].astype(float)
edad_promedio_min = df['Edad Promedio'].min()
edad_promedio_min
```

```python
# País con edad promedio más baja
indice_promedio_min = df['Edad Promedio'].idxmin()
pais_promedio_min = df.iloc[indice_promedio_min]['Pais']
pais_promedio_min
```
